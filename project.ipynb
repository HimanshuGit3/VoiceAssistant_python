```python
import os
import speech_recognition as sr
import pyttsx3
from googlesearch import search
import requests
from bs4 import BeautifulSoup
import smtplib
import pywhatkit as kit

speaker = pyttsx3.init()

def speak(text):
    """Speaks the given text."""
    speaker.say(text)
    speaker.runAndWait()



def web_search(query):
    try:
        # Perform the search and retrieve the first result
        search_results = list(search(query, num=1))  # Use 'num' instead of 'num_results'
        if not search_results:
            return "Sorry, I couldn't find anything relevant."

        # Get the content of the top result
        url = search_results[0]
        response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})  # Use headers to mimic a browser
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract the first few paragraphs of text
        paragraphs = soup.find_all('p')
        if paragraphs:
            answer = ' '.join([para.text for para in paragraphs[:2]])  # Combine first two paragraphs
            return f"I found this on the web: {answer}"
        else:
            return "I couldn't extract useful information from the web page."
    except Exception as e:
        print(f"Web Search Error: {e}")
        return "Sorry, I encountered an error while searching."


def whatsapp_msg():

    phone_number = "+917488113558"  # recipient's phone number
    message = "Hi This message was sent using python"

    kit.sendwhatmsg_instantly(phone_number, message)

def gmail_send():
    
    server = smtplib.SMTP('smtp.gmail.com',587)
    server.starttls()
    server.login('himanshu700.singh@gmail.com', 'wlug bywd genk chdh')
    server.sendmail('himanshu700.singh@gmail.com', 'himanshu308.singh@gmail.com', 'This mail is sent through python')
    print("mail sent successfully")
    return True


def handle_choice(choice):
    choice = choice.lower()  # Normalize input for better matching
    if "notepad" in choice:
        speak("Opening Notepad")
        os.system("notepad")
    elif "youtube" in choice:
        speak("Opening Youtube")
        open_url("https://www.youtube.com", "YouTube")
    elif "whatsapp" in choice and not("send") in choice:
        speak("Opening whatsapp")
        open_url("https://web.whatsapp.com/", "WhatsApp")
    elif "whatsapp" in choice and "send" in choice:
        speak("Sending whatsapp message")
        whatsapp_msg()
    elif "instagram" in choice:
        speak("Opening Instagram")
        open_url("https://www.instagram.com/", "Instagram")
    elif "twitter" in choice or "x" in choice:
        speak("Opening twitter")
        open_url("https://x.com/home", "Twitter")
    elif ("mail" in choice or "gmail" in choice)  and "send" in choice:
        speak("Sending mail")
        gmail_send()
    elif "exit" in choice or "stop" in choice:
        speak("Ok Goodbye! Have a nice day!")
        return False
    else:
        # Handle questions or other unrecognized commands with web search
        speak("Let me search that for you.")
        response = web_search(choice)
        print(f"Web Response: {response}")
        speak(response)
    return True


def open_url(url, description):
    """Opens a URL with a description."""
    speak(f"Opening {description}")
    os.system(f"start {url}")


speak("Hello, I am Ginni, your personal assistant. How can I help you?")

myAI = sr.Recognizer()

while True:
    with sr.Microphone() as mysource:
        try:
            print("\nListening for your command...")
            audio_data = myAI.listen(mysource, timeout=20)
            print("Processing your command...")
            speak("Okay, got it!")
            
            # Recognize and process the command
            choice = myAI.recognize_google(audio_data)
            print(f"Command received: {choice}")
            
            if not handle_choice(choice):
                break
        except sr.UnknownValueError:
            speak("Sorry, I couldn't understand. Could you please repeat?")
        except sr.RequestError:
            speak("It seems there's an issue with the speech service. Please try again later.")
        except Exception as e:
            speak("An error occurred. Please try again.")
            print(f"Error: {e}")


speaker.stop()
```
